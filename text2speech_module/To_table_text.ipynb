```python
import pandas as pd
```

```python
def manual_generation(inputd: dict):

    name = inputd['art_name']
    name = name if name != '' else 'a piece of art'

    author = inputd['author_name']
    author = author if author != '' else 'some artist'

    types = inputd['type']
    name = f'{types} is called {name}' if types != '' else f'is {name}'

    style = inputd['style']
    style = f'It follows the {style} style. ' if style != '' else ''

    date = inputd['date']
    date = f', {date}' if date != '' else ''

    period = inputd['period']
    period = f' in the {period}' if period != '' else ''

    objects = inputd['Tags']
    objects = 'It contains ' + ', '.join(objects) if len(objects) > 0 else ''

    ouput = f"""
        This {name}, created by {author}{period}{date}. {style}{objects}
    """

    return ouput
```

```python
output_list = []

for i, row in df.iterrows():
#     print(i, row)
    name = row['Title']
    name = name if not pd.isna(name) else 'a piece of art'

    author = row['Author']
    author = author if not pd.isna(author) else 'some artist'

    types = row['Genre']
    name = f'{types} is called {name}' if not pd.isna(types) else f'is {name}'

    style = row['Styles']
    style = f'It follows the {style} style. ' if not pd.isna(style) else ''

    date = row['Date']
    date = f', created in {int(date)}' if not pd.isna(date) else ''

#         period = inputd['period']
#         period = f' in the {period}' if period != 'NaN' else ''

    objects = row['Tags']
    if not pd.isna(objects):
        objects = ', '.join(objects.split(',')) if len(objects) > 0 else ''
        objects = f'It contains {objects}' if len(objects) > 0 else ''
    else:
        objects = ''
        
    objects = row['Media']
    if not pd.isna(objects):
        objects = ', '.join(objects.split(',')) if len(objects) > 0 else ''
        objects = f'It create on {objects}' if len(objects) > 0 else ''
    else:
        objects = ''
        

    output = f"""This {name}, It was created by {author}{date}. {style}{objects}"""
#     print(output)
    output_list.append(output)
```

```python
# Create a new DataFrame
output_df = pd.DataFrame({'Index': df.index, 'Text': output_list})
```

```python
import os
import pandas as pd
import pyttsx3

# Assuming you have your DataFrame named 'df'

# Select the range of rows you want (e.g., from index 50 to 100)
start_index = 4458
end_index = 4459

df_subset = output_df.iloc[start_index:end_index]

# Create a directory to save the audio files
directory = '/audio_saved'
os.makedirs(directory, exist_ok=True)

# Initialize the text-to-speech engine
engine = pyttsx3.init("sapi5")
voices = engine.getProperty("voices")[0] 
engine.setProperty('voice', voices)

# Set the speech speed (words per minute)
engine.setProperty('rate', 150)  # Adjust the value as needed (default is around 200)

# Iterate through rows and generate audio files
for index, row in df_subset.iterrows():
    text = row['Text']
    filename = os.path.join(directory, f"{index}.mp3")
    try:
        engine.save_to_file(text, filename)
        engine.runAndWait()
        print(f"Saved {filename}")
    except Exception as e:
        print(f"Error saving {filename}: {e}")
```
